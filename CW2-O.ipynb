```
def normalize_image(image_path):
  image_list = []
  for (j, imagepath) in enumerate(image_path):
    if j%100==0: 
      print(j)
    image = plt.imread(imagepath)
    resized = resize(image, (160,160), mode='reflect')
    resized_image = img_to_array(resized) 
    img = np.expand_dims(resized_image, 0)
    image_list.append(img)
    
  image_list = np.vstack(image_list)
  return image_list
```

```
data = normalize_image(image_path)
data.shape
```

```
data.shape
```

```
data0 = data
```

```
data0.shape
```

```
data_class = np.concatenate([np.ones(270) * i for i in range(14)])
data_class = to_categorical(data_class, 14)
data_class.shape
```

```
#split the data into training data and test data
sss = StratifiedShuffleSplit(n_splits=2, test_size= float(0.1), random_state=1)
for train_index, test_index in sss.split(data0, data_class):
  print("TRAIN:", train_index, "TEST:", test_index)
  X, X_test, y, y_test = data0[train_index], data0[test_index], data_class[train_index], data_class[test_index]
```

```
#split the data into training data and validation data 
sss = StratifiedShuffleSplit(n_splits=2, test_size= float(0.2), random_state=1)
for train_index, test_index in sss.split(X, y):
  print("TRAIN:", train_index, "VALID:", test_index)
  X_train, X_val, y_train, y_valid = X[train_index], X[test_index], y[train_index], y[test_index]
```

```
print(X_train.shape)
print(X_val.shape)
print(y_train.shape)
print(y_valid.shape)
```

```
aug_train = ImageDataGenerator(rotation_range=30,
                               width_shift_range=0.1,
                               height_shift_range=0.1, 
                               shear_range=0.2, 
                               zoom_range=0.2, 
                               horizontal_flip=True, 
                               fill_mode='nearest')
```

#Huấn Luyện Model

```
face_net = load_model('../model/facenet_keras.h5', compile=False)
face_net.load_weights('../weight/facenet_keras_weights.h5')
```

```
rlrop = ReduceLROnPlateau(monitor = 'val_accuracy', patience = 2, mode='max', factor = 0.3, min_lr = 1e-9, verbose = 1)
filepath = 'model.h5'
mdcp = ModelCheckpoint(filepath, monitor='val_accuracy', save_best_only=True, verbose=1)

callbacks = [rlrop, mdcp]
```

```
model2 = Sequential()
model2.add(face_net)
model2.add(Dropout(0.2))
model2.add(Dense(14, activation='softmax',kernel_regularizer= keras.regularizers.l2(0.0002)))
model2.compile(loss= 'categorical_crossentropy',
              optimizer= SGD(momentum =0.9),
              metrics= 'accuracy')
```

```
face_net.trainable = False
```

```
H = model2.fit(aug_train.flow(X_train, y_train, batch_size=17), steps_per_epoch=len(X_train)//17, validation_data=(X_val, y_valid), callbacks=callbacks, epochs=20, verbose=1)
```

```
y_pred = model2.predict(X_test)
```
